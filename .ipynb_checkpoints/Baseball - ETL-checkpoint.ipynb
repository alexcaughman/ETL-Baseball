```python
# identified url
url = 'https://www.baseball-reference.com/teams/ATL/2019-schedule-scores.shtml'
# browser.visit(url)
```

```python
# Use Panda's `read_html` to parse the url
tables = pd.read_html(url)

# time.sleep(25)

tables
```

```python
type(tables)

df = tables[0]
new_df = df[['Date', 'Tm', 'Unnamed: 4', 'W/L']]
new_df
```

```python
cleaned_df = new_df.drop(['Gm#'], axis=0)
cleaned_df = cleaned_df.fillna('')
cleaned_df
```

```python
braves_schedule_df = cleaned_df.rename(columns={"Date":"DATE", "Tm": "PLAYER_TEAM", "Unnamed: 4": "HOME/AWAY"})
braves_schedule_df
```

```python
braves_schedule_df["W/L"].unique()
```

```python
# correcting data 'W-wo', 'L-wo' to W & L
braves_schedule_df.loc[(braves_schedule_df["W/L"] == 'W-wo')]='W'
braves_schedule_df.loc[(braves_schedule_df["W/L"] == 'L-wo')]='L'
braves_schedule_df
```
