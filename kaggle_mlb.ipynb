```python
import pandas as pd
import csv
```

```python
mlb_df = pd.read_csv('kaggle_mlb.csv')
mlb = mlb_df.dropna(how='any')

braves_df = mlb[['PLAYER_NAME', 'PLAYER_TEAM', '2019_HR']]
braves_df
```

```python
braves = braves_df.loc[braves_df["PLAYER_TEAM"] == "Braves", :]
braves
```

```python
braves = braves.rename(columns={"PLAYER_NAME": "player_name", "PLAYER_TEAM": "player_team","2019_HR": "hr_2019"})
braves.head()
```

```python
braves.to_csv('braves_hr.csv', index=False)
```
