```python
# identified url
url = 'https://www.baseball-reference.com/teams/ATL/2019-schedule-scores.shtml'
# browser.visit(url)
```

```python
# Use Panda's `read_html` to parse the url
tables = pd.read_html(url)

# time.sleep(25)

tables
```

```python
type(tables)

df = tables[0]
new_df = df[['Date', 'Tm', 'Unnamed: 4', 'W/L']]
new_df
```

```python
new_df2 = new_df[new_df['Tm'].isin(['ATL'])]
new_df2
```

```python
cleaned_df = new_df2.fillna('')
cleaned_df
```

```python
braves_schedule_df = cleaned_df.rename(columns={"Tm": "PLAYER_TEAM", "Unnamed: 4": "HOMEandAWAY", "W/L": "WorL"})
braves_schedule_df
```

```python
braves_schedule_df["WorL"].unique()
```

```python
braves_schedule_df[["DAY","DATE"]] = braves_schedule_df["Date"].str.split(',', expand = True)
braves_schedule_df
```

```python
del braves_schedule_df['Date']
braves_schedule_df
```

```python
braves_schedule_df.loc[(braves_schedule_df["DATE"] == ' Apr 20 (1)')]=' Apr 20'
braves_schedule_df.loc[(braves_schedule_df["DATE"] == ' Apr 20 (2)')] =' Apr 20'
```

```python
braves_schedule_df[["YEAR","MONTH","DAY"]] = braves_schedule_df["DATE"].str.split(' ', expand = True)
braves_schedule_df
```

```python
braves_schedule_df["YEAR"] = '2019'
braves_schedule_df["YEAR"] =braves_schedule_df["YEAR"].astype(int)
braves_schedule_df["DAY"] =braves_schedule_df["DAY"].astype(int)
month = braves_schedule_df["MONTH"].unique()

braves_schedule_df["MONTH"] = braves_schedule_df['MONTH'].replace(month, [3,4,5,6,7,8,9])
braves_schedule_df
# braves_schedule_df.dtypes
```

```python
# braves_schedule_df['DATE'] =braves_schedule_df[str(braves_schedule_df["DATE"])+str(',2019')]

braves_schedule_df["Year"] = pd.to_datetime(braves_schedule_df[['YEAR', 'MONTH', 'DAY']], format="%Y%m%d")
del braves_schedule_df['DATE']
del braves_schedule_df['DAY']
del braves_schedule_df['MONTH']
del braves_schedule_df['YEAR']
braves_schedule_df
```

```python
braves_schedule_final = braves_schedule_df.rename(columns={"Year": "DATES"})
braves_schedule_final
```

```python
braves_schedule_final.index = np.arange(1, len(braves_schedule_final) + 1)
braves_schedule_final["PLAYER_TEAM"] = braves_schedule_final.replace("ATL", "Braves")
braves_schedule_final.index.names = ['INDEX']
braves_schedule_final = braves_schedule_final.rename(columns={"PLAYER_TEAM": "player_team", "HOMEandAWAY": "homeandaway","WorL": "worl","DATES":"dates"})
braves_schedule_final
```

```python
# Convert to html table
braves_schedule_final.to_csv('brave_2019_schedule.csv')
```
